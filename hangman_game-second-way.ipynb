# simple problem "Guess the word"

```python
import random
```

```python
somewords = [
    'apple', 'bannana', 'mango',
    'strawbwerry', 'orange',
    'grape', 'lemon', 'watermelon',
    'cherry', 'coconut', 'peach', 'figs', 
    'kiwi', 'pineapple'
]
```

## Note here I print random fruit and some things to test the code, you can delete "Print" for these things and the code will do well without any error

```python
print('Guess the word! Hint: word is a name of a fruit')
word = random.choice(somewords)
letterGussed = list('_' * len(word))
# print(letterGussed)
correct_word_list = list(word)
chances = len(word) + 3
wrong_guesses = []
print(word)
while chances > 0:
    print('wrong guesses:',wrong_guesses)
    player_guess = input('Enter a letter to guess: ')
    # print(len(player_guess))

    if player_guess.isalpha() and len(player_guess) > 1:
        print("Enter only one character")
    elif player_guess.isalpha():
        print("all is OK")
    #     print(correct_word_list)
        if player_guess in correct_word_list:
            guess_index = correct_word_list.index(player_guess)
            print(guess_index)
            correct_word_list[guess_index] = '-'
            letterGussed[guess_index] = player_guess
            print(correct_word_list)
            print(f'{letterGussed} ({len(word)} alphabets) ')
            print(f'{"".join(letterGussed)} ({len(word)} alphabets)')
        elif player_guess in letterGussed:
            print("you have already guessed that letter")
            print(f'{letterGussed} ({len(word)} alphabets) ')
        else:
            chances = chances - 1
            print(f"Wrong guessing, you have left with {chances} chances")
            wrong_guesses.append(player_guess)
            print(wrong_guesses)
            if chances <= 0:
                print('You lost! Please try again...')
                print(f"The word was {word}")

    else:
        print("enter only letters")
    if "".join(letterGussed) == word:
        print('Congrats! you have guessed the word correctly')
        break
```

```python
import random as rd

word_bank = {'Person name': ['william', 'peter', 'wallage'],
             'car': ['bmw', 'mercedes', 'toyota']}
word_bank_category = ['Person name', 'car']

category_name = word_bank_category[rd.randint(0, 1)] # display the category name
item_index = rd.randint(0, 2) # display the selection item in each category
correct_word = str(word_bank[category_name][item_index])

correct_word_list = list(correct_word) # check up with user's input
display_list = [char.replace(char, '_') for char in correct_word] # displaying to the users
print('%s: %s(%d alphabets)' % (category_name, display_list, len(display_list)))

wrong_guesses = []
limit_guesses = 0
count = 0

def result():
    if '_' not in display_list:
        print('win!')
        return "stop"

def hangman_drawer():
    global limit_guesses
    if limit_guesses == 1:
        print("   _____ \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
        print("Wrong guess. " + str(limit_guesses) + " guesses remaining\n")
    elif limit_guesses == 2:
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
        print("Wrong guess. " + str(limit_guesses) + " guesses remaining\n")
    elif limit_guesses == 3:
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |    / \ \n"
              "  |      \n"
              "__|__\n")
        print("Wrong guess. " + str(limit_guesses) + " guesses remaining\n")
    elif limit_guesses == 4:
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |    /|\ \n"
              "  |      \n"
              "__|__\n")
        print("Wrong guess. " + str(limit_guesses) + " last guess remaining\n")
    elif limit_guesses == 5:
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |    /|\ \n"
              "  |    / \ \n"
              "__|__\n")
        print("Wrong guess. You are hanged!!!\n")
        print('The word was: "%s"' % correct_word)
        
global limit_guesses, count
while True:
    if result() == "stop":
        break
    print('wrong guesses:',wrong_guesses)
    player_guess = input('Please enter your guess(alphabet): ')
    if player_guess in correct_word_list:
        count += 1
        guess_index = correct_word_list.index(player_guess) # finding the index of the correct alphabet
        correct_word_list[guess_index] = '-'
        display_list[guess_index] = player_guess # replace the dash with the correct alphabet user had input
        print('%s: %s(%d alphabets left)' % (category_name, display_list, len(display_list) - count))
        result()
    else:
        wrong_guesses.append(player_guess)
        limit_guesses += 1
        hangman_drawer()
        result()

```
